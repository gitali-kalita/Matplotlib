```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas=['Marketing','Sales','Development','HR','Customer Support']
budget=[2.5,4,10,1,6]
```

```python
plt.pie(budget, labels=areas, autopct="%0.1f%%")
plt.show()
```

#Scatter Chart

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('salary')
```

#Plot Chart

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]

plt.plot(product_id,online_price)
plt.title("Product Price Comprision")
```

#Bar Chart

```python
year=[2011,2012,2013,2014,2015,2016,2017]
price=[55000,60000,65000,69000,55000,52000,52500]

plt.bar(year,price)
plt.xlabel('year')
plt.ylabel("price(in in INR)")
```
